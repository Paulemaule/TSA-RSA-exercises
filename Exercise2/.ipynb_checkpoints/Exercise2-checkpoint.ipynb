## Task 1.1

```python
# Linear regression to obtain the parameters of the AR model
linear = lambda x, a_0, a_1: a_0 + a_1 * x
estimate_a = lambda x: curve_fit(linear, x[:-1], x[1:])
```

```python
# implementation of a AR(1) model
def AR(a, initial_value, N):
    # a: the parameters of the model, initial_value: the prior/known terms of the time_series, N: the number of additional values to predict
    
    result = np.zeros(N)
    result[0] = a[0] + a[1] * initial_value[-1] #+ np.random.normal(0, 1, 1)
    
    for n in range(1,N):
        result[n] = a[0] + a[1] * result[n-1] #+ np.random.normal(0, 1, 1)
        
    return result

def plot_AR(a, initial_value, N):
    t = np.arange(N)
    plt.plot(t, AR(a, initial_value, N))
```

```python
popt, pcov = estimate_a(data)

print('estimated values for a_0, a_1: \n', popt)
print('estimated covariance matrix of a_0, a_1: \n', pcov)
```

```python
plt.scatter(data[1:], data[:-1])
t = np.arange(-4, 57)
plt.plot(t, linear(t, *popt), color='orange')
plt.title('The estimated parameteter fit with the first return plot')
```

```python
plot_AR(popt, [0,], 50000)
plt.title('The AR(1) model with the estimated parameters and a starting value of 0')
```

```python
plot_AR([0.001, 1], [0,], 50000)
plt.title('The AR(1) model with the optimal parameters [0.001, 1] and a starting value of 0')
```

The parameter $a_1 = 1$ is related to the fact, that this time series is linear in form.

## Task 1.2

```python
# Detrend the time series
popt, pcov = curve_fit(linear, np.arange(len(data)), data)

data_stationary = data - popt[1] * np.arange(len(data))
```

```python
plt.plot(data_stationary)
plt.title('The stationary time series')
```

```python
print(f'The stationary time series has a mean of {np.mean(data_stationary)}')
```

```python
plt.scatter(data_stationary[1:], data_stationary[:-1])
plt.title('The first return plot of the stationary time series')
```

```python
popt, pcov = estimate_a(data_stationary)

print('estimated values for a_0, a_1: \n', popt)
print('estimated covariance matrix of a_0, a_1: \n', pcov)
```

```python
plot_AR(popt, [0,], 50000)
```

With the non stationarity removed the parameters $a_0 = 0.5$ , $a_1 = 0.6$ appear to be the optimal parameters our estimator converges towards.

## Task 1.3

```python
popt, pcov = estimate_a(data)

predicted_1 = np.array([AR(popt, [x,], 1)[-1] for x in data])
predicted_5 = np.array([AR(popt, [x,], 5)[-1] for x in data])
```

```python
t = np.arange(50000)
plt.scatter(t, data, alpha=0.1)
plt.scatter(t, predicted_1, alpha=0.1)
plt.scatter(t, predicted_5, alpha=0.1)
```

```python
error_1 = (data - predicted_1)**2
error_5 = (data - predicted_5)**2
```

```python
plt.scatter(t, error_1)
```

```python
plt.scatter(t, error_5)
```

We can see that both predictions agree relativley well with observation, but the prediction 1 step into the future is of course much better.

# Task 3 - Autocorrelation and partial autocorrelation

```python
data_AR, data_MA = np.loadtxt('Tut2_file3.txt', skiprows=1).T
```

```python
plt.plot(data_AR)
```

```python
plt.plot(data_MA)
```

```python
def autocorrelation(x):
    N = len(x)
    result = np.zeros(int(N/3))
    for dt in range(int(N/3)):
        result[dt] = np.mean((x[dt:] - np.mean(x[dt:])) * (x[:N-dt] - np.mean(x[:N-dt]))) / (np.std(x[dt:]) * np.std(x[:N-dt]))
    
    return result

def partial_autocorrelation(x):
    N = len(x)
    result = np.zeros(int(N/3))
    for dt in range(int(N/3)):
        result[dt] = 0
        
    return result
```

```python
N = 10
plt.scatter(range(N), autocorrelation(data_MA)[:N])
```

```python
N = 1000
plt.scatter(range(N), autocorrelation(data_AR)[:N])
```
