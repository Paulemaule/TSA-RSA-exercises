```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def AR(T = 1000):
    a_0 = np.array([0,0])
    A_1 = np.array([[0.2,0],[-0.2,0.1]])
    A_2 = np.array([[0.1,0],[-0.1,0.1]])
    
    x = np.zeros((T, 2))
    
    x[0] = np.array([0,0])
    x[1] = np.array([0,0])
    
    for t in range(2,T):
        x[t] = a_0 + A_1.dot(x[t-1]) + A_2.dot(x[t-2]) + np.random.normal(0, 0.01, (2))
        
    return x
```

```python
x = AR(1000)
```

```python
plt.figure(figsize=(13,7))
plt.title('The resulting time series')
plt.plot(x[:,0], label='$x_0$')
plt.plot(x[:,1], label='$x_1$')
plt.legend()
```

```python
plt.figure(figsize=(13,7))
plt.title('The resulting time series')
plt.plot(x[:,0], label='$x_0$')
plt.plot(x[:,1], label='$x_1$')
plt.xlim(200,300)
plt.legend()
```

```python
plt.figure(figsize=(7,7))
plt.title('The first return plots')
plt.scatter(x[:-1,0], x[1:,0], label='$x_0$')
plt.scatter(x[:-1,1], x[1:,1], label='$x_1$')
plt.legend()
```

```python
plt.figure(figsize=(7,7))
plt.title('The mixed first return plots')
plt.scatter(x[:-1,1], x[1:,0], label='$x_{1,t-1} \leftrightarrow x_{0,t}$')
plt.scatter(x[:-1,0], x[1:,1], label='$x_{0,t-1} \leftrightarrow x_{1,t}$')
plt.scatter(x[:-1,0], x[1:,1], label='$x_{0,t-1} \leftrightarrow x_{1,t}$')
plt.scatter(x[:-1,0], x[1:,1], label='$x_{0,t-1} \leftrightarrow x_{1,t}$')
plt.legend()
```
